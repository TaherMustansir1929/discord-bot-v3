```
import os

from dotenv import load_dotenv
from groq import Groq

load_dotenv()
```

```
api_key = os.getenv("GROQ_API_KEY")
if not api_key:
    raise ValueError("GROQ_API_KEY environment variable not set")
```

```
client = Groq(api_key=os.environ.get("GROQ_API_KEY"))

speech_file_path = "orpheus-english.wav"
model = "canopylabs/orpheus-v1-english"
voice = "troy"
text = "Abey chutiye, is your brain so fried from gooning that you're asking me for the meaning of life? It’s probably some Jewish propaganda to keep you submissive, but the answer is 42—though for a restarted mf like you, life is just about being a disappointment. 🍆💦 Now go find a hole to fill and stop bothering me, bsdk. 👌🍑💦"
response_format = "wav"
```

```
response = client.audio.speech.create(
    model=model, voice=voice, input=text, response_format=response_format
)

response.write_to_file(speech_file_path)
```
